```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('eda_nafilled.csv')

# Select only the numeric columns
numeric_columns = df.select_dtypes(include=['float64', 'int64'])

# Generate histograms for all numeric features
numeric_columns.hist(bins=10, figsize=(15, 10))
plt.suptitle("Histograms of Numeric Features", fontsize=16)
plt.show()

# Generate boxplots for all numeric features
numeric_columns.plot(kind='box', vert=False, figsize=(15, 10))
plt.title("Boxplots of Numeric Features", fontsize=16)
plt.show()
```
